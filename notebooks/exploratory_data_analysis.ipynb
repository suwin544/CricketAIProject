```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline
```

```python
# Define paths to raw data folders
t20_path = "../data/raw/t20s_csv2"
ipl_path = "../data/raw/ipl_csv2"
```

```python
# Load only match files — skip _info.csv files
t20_files = [os.path.join(t20_path, f) for f in os.listdir(t20_path) if f.endswith('.csv') and not f.endswith('_info.csv')]
ipl_files = [os.path.join(ipl_path, f) for f in os.listdir(ipl_path) if f.endswith('.csv') and not f.endswith('_info.csv')]

# Read and combine
t20_df = pd.concat((pd.read_csv(file) for file in t20_files), ignore_index=True)
ipl_df = pd.concat((pd.read_csv(file) for file in ipl_files), ignore_index=True)
```

```python
print("T20 Shape:", t20_df.shape)
print("IPL Shape:", ipl_df.shape)

t20_df.head()
```

```python
# List all available columns
print("T20 Columns:", t20_df.columns.tolist())
print("IPL Columns:", ipl_df.columns.tolist())
```

```python
# Check for missing/null values
print("Missing values in T20 dataset:")
print(t20_df.isnull().sum())

print("\nMissing values in IPL dataset:")
print(ipl_df.isnull().sum())
```

```python
# Basic statistical summary
print(t20_df.describe())

# Adjusted key column names to match dataset
print("\nTop 5 Strikers:", t20_df['striker'].value_counts().head())
print("Top 5 Bowlers:", t20_df['bowler'].value_counts().head())
print("Dismissal Types:", t20_df['wicket_type'].value_counts())
```

```python
# Save processed datasets
t20_df.to_csv("../data/processed/t20s_combined.csv", index=False)
ipl_df.to_csv("../data/processed/ipl_combined.csv", index=False)
```

```python
## Match Frequency and Innings Distribution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Match frequency by season
plt.figure(figsize=(12, 5))
sns.countplot(data=t20_df, x='season', order=sorted(t20_df['season'].astype(str).unique()))
plt.title("Number of T20 Matches per Season")
plt.xticks(rotation=45)
plt.show()

# Innings distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=t20_df, x='innings')
plt.title("Distribution of Innings")
plt.show()
```

```python
# Top 10 venues by match count
plt.figure(figsize=(12, 5))
sns.countplot(data=t20_df, y='venue', order=t20_df['venue'].value_counts().head(10).index)
plt.title("Top 10 Venues by Match Count")
plt.xlabel("Number of Matches")
plt.ylabel("Venue")
plt.show()

# Distribution of runs scored off the bat
plt.figure(figsize=(10, 5))
sns.histplot(t20_df['runs_off_bat'], bins=30, kde=True)
plt.title("Distribution of Runs Scored Off the Bat")
plt.xlabel("Runs per Ball")
plt.ylabel("Frequency")
plt.show()
```

```python
# Top batting teams by total runs scored
plt.figure(figsize=(10, 5))
t20_df.groupby('batting_team')['runs_off_bat'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Batting Teams by Total Runs")
plt.ylabel("Total Runs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top bowling teams by total wickets
plt.figure(figsize=(10, 5))
t20_df['wicket'] = t20_df['wicket_type'].notnull().astype(int)  # 1 if a wicket type exists, else 0
t20_df.groupby('bowling_team')['wicket'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Bowling Teams by Total Wickets")
plt.ylabel("Wickets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Dismissal Types
plt.figure(figsize=(8, 4))
t20_df['wicket_type'].value_counts().plot(kind='bar', color='crimson')
plt.title("Distribution of Dismissal Types")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Extras by Bowling Team
plt.figure(figsize=(10, 5))
t20_df.groupby('bowling_team')['extras'].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='green')
plt.title("Top 10 Teams by Total Extras Conceded")
plt.ylabel("Extras")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Add over column from 'ball' (e.g., 10.2 → 10)
t20_df['over'] = t20_df['ball'].astype(str).str.extract(r'(\d+)\.').astype(float)

# Average runs per over
plt.figure(figsize=(10, 5))
t20_df.groupby('over')['runs_off_bat'].mean().plot(kind='line', marker='o')
plt.title("S")
plt.xlabel("Over")
plt.ylabel("Runs Off the Bat")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Wickets per over
plt.figure(figsize=(10, 5))
t20_df.groupby('over')['wicket'].sum().plot(kind='bar', color='darkred')
plt.title("Wickets Taken per Over")
plt.xlabel("Over")
plt.ylabel("Wickets")
plt.tight_layout()
plt.show()
```

```python
# Numeric correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(t20_df[['runs_off_bat', 'wides', 'noballs', 'byes', 'legbyes', 'penalty', 'wicket']].corr(), 
            annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
